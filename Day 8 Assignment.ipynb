### Question 1: Recreate “NB 21 map, filter, lambda.ipynb” to try different variations or perform different calculations from the ones that are in the NB

```python
## map() - It applies the given function to all iterables and returns a new list.
## SYNTAX: map(function, iterables)

list1=["apple", "banana", "cherry"]
x1=list(map(str.upper,list1))
print("Uppercase of the input: ", x1)


def addtn(x,y):
    return x+y
x2=list(map(addtwo,[i for i in range(1,20)],[j for j in range(21,40)]))
print(x2)

## Lambda functions -  Lambda expressions are anonymous or nameless functions.
## syntax: lambda arguments: expression

x3=lambda x:x+2
print("Adding two numbers: ", x3(15))

x4=lambda x,y:x+y
print("Addition of two numbers: ",x4(11,21))

x5=list(map(lambda x:x**2,[i for i in range(1,10)]))       ## using map function with lambda function
print(x5)

## filter() - It is used to filter the given iterables with the help of another function passed as an argument to test all the elements to be True or False.
## SYNTAX: filter(function, iterables)

def morethan21(y):
    if y>21:
        return y

x6=filter(morethan21,[i for i in range(1,50)])
print("Filtered numbers in a list which are greater than 21: ", list(x6))

x7=filter(lambda x:x%7==0,[i for i in range(1,50)])
print("Filtered numbers in a list which are divisible by 7: ", list(x7))
```
